```python
import os, glob, numpy as np
from quantem.widget import IO, Show4DSTEM
```

```python
DATA_DIR = os.path.expanduser("~/Desktop/_Research/_quantem-workspace/quantem.widget/notebooks/io/koreansampleh5")
masters = sorted(glob.glob(os.path.join(DATA_DIR, "*_master.h5")))
assert masters, f"No *_master.h5 found in {DATA_DIR}"
for m in masters:
    print(os.path.basename(m))

result = IO.arina_file(masters[0], det_bin=4)
print(f"\n{result.data.shape}, {result.data.dtype}")
```

```python
w = Show4DSTEM(result)
w
```

```python
w_icom = Show4DSTEM(result, title="iCOM")
w_icom.roi_mode = "icom"
w_icom
```

```python
w_mag = Show4DSTEM(result, title="COM magnitude")
w_mag.roi_mode = "com_mag"
w_mag
```

```python
w_all = Show4DSTEM(result, title="COM/DPC stats")
sr, sc = w_all.shape_rows, w_all.shape_cols
for mode in ["com_x", "com_y", "com_mag", "icom", "dcom", "curl"]:
    w_all.roi_mode = mode
    vi = np.frombuffer(w_all.virtual_image_bytes, dtype=np.float32).reshape(sr, sc)
    print(f"  {mode:8s}: [{vi.min():.4f}, {vi.max():.4f}], std={vi.std():.4f}")
w_all.roi_mode = "icom"
w_all
```

```python
w_dc = Show4DSTEM(result, title="dCOM")
w_dc.roi_mode = "dcom"
w_dc
```

```python
w_curl = Show4DSTEM(result, title="Curl")
w_curl.roi_mode = "curl"
w_curl
```

```python
w_x = Show4DSTEM(result, title="COM-X")
w_x.roi_mode = "com_x"
w_x
```

```python
w_y = Show4DSTEM(result, title="COM-Y")
w_y.roi_mode = "com_y"
w_y
```

```python
for w in [w, w_icom, w_mag, w_all, w_dc, w_curl, w_x, w_y]:
    w.free()
del result
```
