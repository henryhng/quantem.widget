# IO.folder

Read all files in a folder as a stack with `IO.folder(path)`. Supports `file_type` filter and `recursive` flag.

```python
from quantem.widget import IO, Show3D, profile
profile()
```

## TIFF folder

```python
result = IO.folder("/Users/macbook/data/20251216_Korea_Sample_C1/", file_type="tif")
print(result)
Show3D(result, title="Korean Sample Virtual Images")
```

## EMD folder

```python
result = IO.folder("/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/", file_type="emd", shape=(1024, 1024))
print(result)
Show3D(result, title="Korean D1 STEM HAADF Series")
```

## DM3 folder

```python
result = IO.folder(
    "/Users/macbook/data/20251209_KoreanSample/C1_TEM_Bob/Bob_Lee_Sangjoon/",
    file_type="dm3",
)
print(result)
Show3D(result, title="Co3O4 TEM Series")
```

## Recursive folder

```python
result = IO.folder(
    "/Users/macbook/data/20251209_KoreanSample/C1_TEM_Bob/Bob_Lee_Sangjoon/",
    file_type="dm3",
    recursive=True,
)
print(result)
```
