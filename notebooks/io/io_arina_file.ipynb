# IO.arina_file — single file

Load a single arina master file and display with Show4DSTEM.

- `det_bin` bins the detector (e.g. `det_bin=4` turns 192×192 → 48×48)
- `det_bin="auto"` picks the smallest factor that fits in memory
- Backend is auto-detected (CUDA on NVIDIA, MPS on macOS, CPU fallback)

```python
master_path = '/home/bobleesj/data/Korean_C1/KoreanSampleC1Bob_06_master.h5'
dataset = IO.arina_file(master_path, det_bin=4)
print(dataset)
```

```python
viewer = Show4DSTEM(dataset)
```
