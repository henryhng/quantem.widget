# IO.arina_folder — load all scans in a folder

Finds every `*_master.h5` in a directory, loads each with CUDA,
and stacks into a 5D dataset `(n_files, scan_r, scan_c, det_r, det_c)`.

- `pattern` filters by filename (case-insensitive)
- `recursive=True` searches subdirectories
- `max_files` caps how many files are loaded (default 50)

## Benchmark network throughput

Before loading a large dataset over a network mount, check read speed with `IO.benchmark()`.
This reads a small sample (64 MB by default) and reports throughput + estimated full load time.

```python
IO.benchmark('/home/bobleesj/data/Korean_C1/')
```

## Single folder

```python
dataset = IO.arina_folder('/home/bobleesj/data/Korean_C1/', det_bin=4)
print(dataset)
```

## Multiple files → 5D stack

Pass a list of master files to `IO.arina_file` to get a 5D dataset.
Use the frame slider in Show4DSTEM to switch between scans.

```python
master_files = [
    '/home/bobleesj/data/Korean_C1/KoreanSampleC1Bob_06_master.h5',
    '/home/bobleesj/data/geroge/gold-10/gold_10_master.h5',
]
stack = IO.arina_file(master_files, det_bin=4)
print(stack)
```

```python
# Show4DSTEM expects 4D — select one scan from the 5D stack
viewer = Show4DSTEM(stack.data[0])
viewer
```

```python
viewer.free()
del stack, viewer; free_gpu()
```
