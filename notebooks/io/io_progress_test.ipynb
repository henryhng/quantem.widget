# IO Progress Bars & Benchmark

Verify tqdm progress bars appear during multi-file loads and `IO.benchmark()` reports throughput.

```python
from quantem.widget import IO, Show2D, Show3D, profile
profile()
```

## Benchmark

```python
IO.benchmark("/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/")
```

## IO.folder — EMD (4 files)

```python
# This folder has mixed sizes (2048x2048 + 1024x1024) — shows helpful error
try:
    IO.folder("/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/", file_type="emd")
except ValueError as e:
    print(e)
print()
# Use shape= to filter — prints how many files were kept/dropped
result = IO.folder("/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/", file_type="emd", shape=(1024, 1024))
print(result)
```

## IO.folder — DM3 (5 files)

```python
result = IO.folder("/Users/macbook/data/20251209_KoreanSample/C1_TEM_Bob/Bob_Lee_Sangjoon/", file_type="dm3")
print(result)
Show3D(result, title="Co3O4 TEM Series")
```

## IO.folder — TIFF (40 files)

```python
result = IO.folder("/Users/macbook/data/20251216_Korea_Sample_C1/", file_type="tif")
print(result)
Show3D(result, title="Korean Sample Virtual Images")
```

## IO.file — list of EMD files

```python
# IO.file with mixed sizes — same shape= API works here too
try:
    IO.file([
        "/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/0004 - 330 kx STEM HAADF.emd",
        "/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/0005 - 3.70 Mx STEM HAADF.emd",
    ])
except ValueError as e:
    print(e)
```

```python
# IO.file with shape= filter — load same-size files from a mixed list
result = IO.file([
    "/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/0005 - 3.70 Mx STEM HAADF.emd",
    "/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/0006 - 5.20 Mx STEM HAADF.emd",
    "/Users/macbook/data/20251209_KoreanSample/D1_STEM_Karen/0007 - 3.70 Mx STEM HAADF.emd",
])
print(result)
Show3D(result)
```
