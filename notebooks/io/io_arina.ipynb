# IO.arina_file → Show4DSTEM

Load Arina 4D-STEM data and display with `Show4DSTEM`.

- **macOS (Apple Silicon)**: uses Metal GPU decompression (fastest)
- **Windows / Linux / any platform**: uses h5py + hdf5plugin CPU fallback (works everywhere)

Backend is auto-detected — no configuration needed.

```python
from quantem.widget import IO, Show4DSTEM
```

## Load arina master file

`det_bin` bins the detector by that factor (e.g. `det_bin=4` turns 192×192 → 48×48).
Use `det_bin="auto"` to let the loader pick based on available memory.

```python
master_path = "/Users/macbook/data/20251216_Korea_Sample_C1/KoreanSampleC1Bob_00_master.h5"
result = IO.arina_file(master_path, det_bin=4)
print(result)
```

## Display with Show4DSTEM

Pass the `IOResult` directly — title and metadata are extracted automatically.

```python
w = Show4DSTEM(result)
w
```

## Explore

Click on the virtual image to navigate scan positions. Use presets (BF, ABF, ADF) for quick virtual imaging modes.

```python
w.summary()
```

## Cleanup

Release GPU memory when done (important for large datasets in notebook sessions).

```python
w.free()
del result
```
