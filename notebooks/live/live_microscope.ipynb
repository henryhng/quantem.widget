# Live Acquisition Viewer

Real-time viewer for incoming microscope data. Change the folder path to your acquisition directory.

```python
from quantem.widget import Live

# Point this to your acquisition folder
FOLDER = "/Users/macbook/data/bob/20260409_gold_drift_v3"

live = Live.watch(FOLDER, pattern="*.emd", interval=2)
live
```

## Manual push (alternative)

If you prefer to push frames manually instead of watching a folder:

```python
from quantem.widget import IO, Live

# Load existing data
result = IO.folder("/Users/macbook/data/bob/20260409_gold_drift_v3", file_type="emd", shape=(4096, 4096))

# Create Live and push frames one at a time
live2 = Live(title="Gold Drift Manual")
for i in range(result.data.shape[0]):
    live2.push(result.data[i], label=result.labels[i])
live2
```
