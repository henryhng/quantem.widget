# Live — Real Microscopy Data

Browse real EM data files with the Live widget. Uses `from_folder()` to bulk-load all files, then navigate with arrow keys or click thumbnails.

```python
from quantem.widget import Live, IO, profile
profile()
```

## STEM HAADF (Velox EMD, 1024×1024)

Browse all STEM HAADF images from a Korean sample session. Click thumbnails to inspect, use ← → keys to navigate.

```python
w_emd = Live.from_folder(
    "/Users/macbook/data/korean/20251209_KoreanSample/D1_STEM_Karen/",
    pattern="*.emd",
    title="STEM HAADF (Velox)",
    cmap="inferno",
    log_scale=True,
)
w_emd
```

## TEM Images (DM3, 4096×4096)

High-resolution TEM images. These are 4K — the widget handles them via incremental push (64 MB per frame).

```python
w_dm3 = Live.from_folder(
    "/Users/macbook/data/korean/20251209_KoreanSample/C1_TEM_Bob/Bob_Lee_Sangjoon/",
    pattern="*.dm3",
    title="TEM 4K (DM3)",
    cmap="gray",
)
w_dm3
```

## Virtual Images (TIFF)

4D-STEM derived virtual BF/ADF/HAADF images. Small files, fast loading.

```python
w_tiff = Live.from_folder(
    "/Users/macbook/data/korean/20260304_korean_c1/",
    pattern="*.tif",
    title="Virtual Images (TIFF)",
    cmap="viridis",
)
w_tiff
```

## Mixed push from IO.file

Push individual files from different locations — mixed formats and sizes in one viewer.

```python
w_mixed = Live(title="Mixed Formats")
# EMD
w_mixed.push(IO.file("/Users/macbook/data/korean/20251209_KoreanSample/D1_STEM_Karen/0007 - 3.70 Mx STEM HAADF.emd"), label="EMD HAADF 1K")
# DM3
w_mixed.push(IO.file("/Users/macbook/data/korean/20251209_KoreanSample/C1_TEM_Bob/Bob_Lee_Sangjoon/final_0001.dm3"), label="DM3 TEM 4K")
# TIFF
w_mixed.push(IO.file("/Users/macbook/data/korean/20260304_korean_c1/korean_sample_c1_zone_axis003_Virtual00_8bit.tif"), label="TIFF Virtual")
print(f"Mixed: {w_mixed.n_frames} frames, labels={w_mixed.labels}")
w_mixed
```
