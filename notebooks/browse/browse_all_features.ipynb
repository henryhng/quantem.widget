## Load selected arina master → Show4DSTEM

## Load single image → Show2D

Navigate to `D1_STEM_Karen/` and select an `.emd` file.

```python
if browser.path:
    r = browser.load()
    Show2D(r.data, title=r.title, pixel_size=r.pixel_size, show_fft=True)
else:
    print("Select a file above first.")
```

## Multi-file selection → Show3D stack

```python
multi = Browse(root="/Users/macbook/data", selection_mode="files", filter_exts=[".emd", ".dm3", ".tif"], title="Select Multiple Files")
multi
```

```python
if multi.selected:
    r = IO.file(list(multi.selected))
    Show3D(r.data, labels=r.labels, title=f"{len(multi.selected)} files")
else:
    print("Select multiple files above (checkboxes), then re-run this cell.")
```

## Folder selection

```python
folder_browser = Browse(root="/Users/macbook/data", selection_mode="folder", title="Select Data Folder")
folder_browser
```

```python
if folder_browser.path:
    r = IO.folder(folder_browser.path)
    Show3D(r.data, labels=r.labels, title=r.title)
else:
    print("Select a folder above, then re-run this cell.")
```
